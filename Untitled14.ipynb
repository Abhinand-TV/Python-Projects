```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#to upload file in system
from google.colab import files
uploaded = files.upload()


```

```python
#reading files
df= pd.read_csv("Books_Data_Clean.csv")
```

```python
df.head()
```

```python
"Summary statistics of numerical columns"
df.describe()
```

```python
#to fix negative outlier in publishing year
df=df[df["Publishing Year"]>1900]
```

```python
#to check if we have NA Values in df
df.isna().sum()
```

```python
#to drop rows with no book name
df.dropna(subset="Book Name",inplace=True)
```

```python
#to check if there are any duplicate rows
df.duplicated().sum()
```

```python
#check unique values
df.nunique()
```

```python
#to plot the histogram
plt.hist(df["Publishing Year"])
plt.xlabel("Publishing Year")
plt.ylabel("Frequency")
plt.title("Distribution of publishing Year")
plt.show()
```

```python
#to plot the bar plot
genre_counts=df["genre"].value_counts()
plt.bar(genre_counts.index,genre_counts.values)
plt.xlabel("Genre")
plt.ylabel("Number of books")
plt.title("Distribution of Genre")
plt.show()
```

```python
df.groupby("Author")["Book_average_rating"].mean()
```

```python
#boxplot
sns.boxplot(x="genre",y="Book_ratings_count",data=df)
plt.xlabel("Genre")
plt.ylabel("Book  ratings")
plt.title("Box plot of Book ratings counts for each Genre")
plt.show()
```

```python
plt.scatter(df["sale price"],df["units sold"])
plt.xlabel("sale price")
plt.ylabel("units sold")
plt.title("Scatter plot of sale price vs units sold")
plt.show()
```

```python
df["language_code"].value_counts().sort_values(ascending=True)
```

```python
language_counts=df["language_code"].value_counts()
```

```python
plt.pie(language_counts, labels= language_counts.index, startangle=90)
plt.title("Distribution of Languages")
plt.show()
```

```python
df.groupby("Publisher ")["publisher revenue"].sum().sort_values(ascending=False)
```

```python
df.groupby("Author_Rating")["Book_ratings_count"].mean().sort_values(ascending=False)
```

```python
df.groupby("language_code").size().sort_values(ascending=False)
```

```python
df.groupby("Author_Rating")["Book_ratings_count"].max().sort_values(ascending=False)
```

```python
plt.scatter(df["Book_average_rating"],df["Book_ratings_count"])
plt.xlabel("Book_average_rating")
plt.ylabel("Book_ratings_count")
plt.title("Scatter plot for book average rating vs ratings count")
plt.show()
```

```python
total_gross_sales_by_author=df.groupby("Author")["gross sales"].sum()
```

```python
total_gross_sales_by_author.sort_values(ascending=False).head(10).plot(kind="bar")
plt.xlabel("Author")
plt.ylabel("Total Gross Sales")
plt.title("Total Gross Sales by Author")
plt.show()
```

```python
sns.boxplot(x="Author_Rating",y="units sold",data=df)
plt.xlabel("Author_Rating")
plt.ylabel("units sold")
plt.title("Box plot of units sold for each Author Rating")
plt.show()
```

```python
df.groupby("Publishing Year")["units sold"].sum().plot(kind="line",marker="o")
plt.xlabel("Publishing Year")
plt.ylabel("Total units sold")
plt.title("Total units sold over time")
plt.show()
```
